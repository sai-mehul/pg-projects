```python
import pandas as pd 
import numpy as np 
from random import Random
import random
```

```python
#we cant fit time series data less than 50 observations (i.e. small data set) t becomes difficult to accurately capture underlying patterns like trends and seasonality
#we then simulate it from 
random.seed(568)
population = [random.normalvariate(10, 5) for _ in range(1000)]
population = [x for x in population if x >= 0]# lets say these are amount of cars that enter within a period 
days = [_ for _ in range(len(population))]
```

```python
population = [int(car) for car in population]
print(population)
print(days)
```

```python
dict_ = {'Day': days, 'Cars': population}
data = pd.DataFrame(dict_, index=days)
```

```python
data
```

```python
sample = data.sample(n=40)
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf
```

```python
model = ARIMA(sample['Cars'],order=(1,0,0))
```

```python
fitted_Model = ARIMA.fit(model)
```
